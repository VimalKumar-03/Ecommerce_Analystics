# Problem Statement

```python
# Importing libraries for numerical calculation and data manipulation and Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("E-Commerce Analytics Vendor.csv")
df
```

# Data Understanding and Cleaning

```python
df.head(10)
```

```python
df.shape
```

# Cleaning Column structure

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df.columns.tolist()
```

# Remove Completely Empty Keywords

```python
df = df[df["search_keyword"].notna()]
df.shape
```

# Cleaning Search Keyword Text

```python
df["search_keyword"] = (
    df["search_keyword"]
    .str.lower()
    .str.strip()
)
```

# Cleaning Date Column

```python
df["search_date"].head(3)
```

```python
# Removing From GMT till end
df["search_date_clean"] = df["search_date"].str.replace(
    r" GMT.*", "", 
    regex=True
)
```

```python
df["search_date_clean"].head(3)
```

```python
# Converting to a proper DateTime Format
df["search_date_dt"] = pd.to_datetime(
    df["search_date_clean"],
    format="%a %b %d %Y %H:%M:%S",
    errors="coerce"
)
```

```python
print(df["search_date_dt"])
```

```python
df["search_date"] = df["search_date_dt"]
df = df.drop(columns=["search_date_clean", "search_date_dt"])
```

```python
df["week"] = df["search_date"].dt.to_period("W").astype(str)
```

```python
df[["search_keyword", "search_date"]].head(10)
```

```python
df["week"].head(10)
```

# KPI 1: Zero-Result Searches (Metric #3)

```python
zero_result_kpi = (
    df[df["total_results"] == 0]
    .groupby("week")
    .size()
    .reset_index(name="zero_result_searches")
)

zero_result_kpi
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))

sns.lineplot(
    data=zero_result_kpi,
    x="week",
    y="zero_result_searches",
    marker="o"
)

plt.title("Weekly Zero Result Searches")
plt.xlabel("Week")
plt.ylabel("Zero Result Searches")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# KPI-2: Low-Result Searches (< 3 results)

```python
low_result_kpi = (
    df[df["total_results"] < 3]
    .groupby("week")
    .size()
    .reset_index(name="low_result_searches")
)

low_result_kpi
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))

sns.lineplot(
    data=low_result_kpi,
    x="week",
    y="low_result_searches",
    marker="o"
)

plt.title("Weekly Low Result Searches")
plt.xlabel("Week")
plt.ylabel("Low Result Searches")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# KPI-3: Search Fail Rate

```python
weekly_total = (
    df.groupby("week")
      .size()
      .reset_index(name="total_searches")
)

weekly_zero = (
    df[df["total_results"] == 0]
    .groupby("week")
    .size()
    .reset_index(name="zero_searches")
)

weekly_low = (
    df[df["total_results"] < 3]
    .groupby("week")
    .size()
    .reset_index(name="low_searches")
)

fail_rate = weekly_total.merge(weekly_zero, on="week", how="left") \
                        .merge(weekly_low, on="week", how="left")

fail_rate[["zero_searches", "low_searches"]] = fail_rate[["zero_searches", "low_searches"]].fillna(0)

fail_rate["search_fail_rate"] = (
    (fail_rate["zero_searches"] + fail_rate["low_searches"])
    / fail_rate["total_searches"]
)

print(fail_rate)
```

The search fail rate is very high across all three weeks, staying between 38% and 41%. 
This means that around 4 out of 10 user searches are not getting proper results. 
Even though the total search volume increased heavily in the second week, the failure rate remained high.
This indicates serious issues in product availability or search relevance that can directly impact sales and user experience.

# KPI-4: Trending Keywords (Week-over-Week Growth)

```python
weekly_keyword_volume = (
    df.groupby(["week", "search_keyword"])
      .size()
      .reset_index(name="search_count")
      .sort_values(["search_keyword", "week"])
)

weekly_keyword_volume.head(10)
```

# Calculate Week-over-Week (WoW) Growth

```python
weekly_keyword_volume["prev_week_count"] = (
    weekly_keyword_volume
    .groupby("search_keyword")["search_count"]
    .shift(1)
)

weekly_keyword_volume["wow_growth"] = (
    (weekly_keyword_volume["search_count"] - weekly_keyword_volume["prev_week_count"])
    / weekly_keyword_volume["prev_week_count"]
)
```

```python
trending_keywords = (
    weekly_keyword_volume
    .dropna(subset=["wow_growth"])
    .sort_values(by="wow_growth", ascending=False)
)

trending_keywords.head(10)
```

# KPI-8: Price Intent (User Price Preference)

```python
df["min_price"] = pd.to_numeric(df["min_price"], errors="coerce")
df["max_price"] = pd.to_numeric(df["max_price"], errors="coerce")
```

```python
# Creating Average Price Intent per Search
df["avg_price_intent"] = df[["min_price", "max_price"]].mean(axis=1)
```

```python
# Create Price Buckets (Low / Medium / High)
df["price_segment"] = pd.cut(
    df["avg_price_intent"],
    bins=[0, 300, 800, 2000, 100000],
    labels=["Low Budget", "Mid Budget", "High Budget", "Premium"]
)
```

```python
# Price Intent Distribution
price_intent_kpi = (
    df.groupby("price_segment")
      .size()
      .reset_index(name="search_count")
      .sort_values(by="search_count", ascending=False)
)

price_intent_kpi
```

```python
sns.set_theme(style="whitegrid", palette="Set2")

plt.figure(figsize=(8, 5))

sns.barplot(
    data=price_intent_kpi,
    x="price_segment",
    y="search_count"
)

plt.title("Search Distribution by Price Segment")
plt.xlabel("Price Segment")
plt.ylabel("Search Count")
plt.tight_layout()
plt.show()
```
